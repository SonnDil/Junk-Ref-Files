```python
import fitz  #pymupdf
import pandas as pd

def extract_text_from_rect_area(page, bbox):
    rect = fitz.Rect(bbox) #create a rect from bbox coord
    text = page.get_text("text", clip = rect)
    return text.replace("\n", " ")

def extract_column_data(page, x_start_range,x_end_range,y_start_first,num_rows, y_start_diff=3):
    data_frames = []
    for x_start, x_end in zip(x_start_range, x_end_range):
        data = {'Text':[]}
        for i in range(num_rows):
            y_start = y_start_first + i * y_start_diff
            y_end = y_start + y_start_diff
            bbox = (x_start,y_start,x_end,y_end)
            text_in_area = extract_text_from_rect_area(page, bbox)
            data['Text'].append(text_in_area)
        data_frames.append(pd.DataFrame(data))
    return pd.concat(data_frames, axis = 1)


###########################################################
#Example usage
#give the file path and the page number want to extract
doc = fitz.open("Vytalogy Financial Results v4_Oct_OCR.pdf")
page = doc[2]

# change the coordinates according to the format and
# can use this to identify coordinates --> page.get_text("blocks")
x_start_range = [18.2, 117,145,175,206,  239,256,283,311,  330,365,391,421,  450,476,501,526  ,551,581,609,636,  662,686,711,736,  760,786,811,841,  868,889,913,936]
x_end_range =  [114, 144,175,202,226,  255,279,305,325,  360,390,420,445,  475,500,525,550  ,578,608,635,655,  685,710,735,756,  785,810,840,860,  884,910,935,975]

# top left y coordinate of the table
y_start_first= 1

# play around with the num rows since we get duplicates
num_rows = 84 

# gap between the rows of the table (trial and error)
# higher values get merged rows, so better to go with low difference
y_start_diff = 6.3


merged_df = extract_column_data(page, x_start_range,x_end_range,y_start_first,num_rows, y_start_diff)
first_row = merged_df.iloc[0]
merged_df.drop(0,inplace=True)
merged_df.columns = first_row
merged_df


#get the extracted table as a CSV file and use this to preprocess (will include partial duplicates)
#merged_df.to_csv("Vytalogy Financial Results v4_Oct_OCR_pg3.csv", header= True, index=None)
```
